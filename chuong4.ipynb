#Thực hành 2: Vẽ các đồ thị

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
fs = [1,2,4]
alltime = np.linspace(0,2,200)
t = alltime[:100]
```

```python
from google.colab import drive
drive.mount('/drive')
```

```python
for f in fs:
  y = np.sin(2*np.pi*f*t)
  plt.plot(t,y,label='Hz'.format(f))
plt.legend()
plt.savefig('/drive/My Drive/toanchuong4/basic_python.pdf')
```
